```python
mapping = {'Female': 1,
           'Male': 0}

adult_df['sex'] = adult_df['sex'].map(mapping)
```

```python
adult_df['income'] = LabelEncoder().fit_transform(adult_df['income'].tolist())
```

```python
adult_df
```

```python
adult_df = pd.get_dummies(adult_df)
```

```python
adult_df.columns
```

```python
X = adult_df.drop(columns='income').values
X
```

```python
y = adult_df['income'].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
parameters = {'n_neighbors': np.arange(1, 15)}
```

```python
knc = KNeighborsClassifier()
clf = GridSearchCV(knc, parameters)
clf.fit(X_train, y_train)
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
